`reshape` меняет геометрию массива, но не число элементов.

**Сигнатура:** `arr.reshape(shape, order='C')`.


```python
x = np.arange(12)
x2 = x.reshape(3, 4)
print("x2:\n", x2)

assert x.size == x2.size
assert x2.shape == (3, 4)
```

## 2) `transpose`, `.T`, `swapaxes`, `moveaxis`

- `transpose`/`.T` переставляют оси;
- `swapaxes` меняет местами две заданные оси;
- `moveaxis` переносит ось в новую позицию.


```python
m = np.arange(6).reshape(2, 3)
mt = m.T

cube = np.arange(24).reshape(2, 3, 4)
sw = np.swapaxes(cube, 0, 2)
mv = np.moveaxis(cube, 0, -1)

print("m:\n", m)
print("m.T:\n", mt)
print("swapaxes shape:", sw.shape)
print("moveaxis shape:", mv.shape)

assert mt.shape == (3, 2)
assert sw.shape == (4, 3, 2)
assert mv.shape == (3, 4, 2)
```

## 3) `ravel` vs `flatten` и идея `view/copy`

- `ravel` обычно возвращает view (когда возможно);
- `flatten` возвращает копию.


```python
a = np.arange(6).reshape(2, 3)
av = a.ravel()
ac = a.flatten()

a[0, 0] = 999

print("a:\n", a)
print("ravel:", av)
print("flatten:", ac)

assert av[0] == 999
assert ac[0] == 0
```

## 4) `expand_dims` и `squeeze`

- `expand_dims(a, axis)` добавляет новую ось размера 1;
- `squeeze(a, axis=None)` удаляет оси размера 1.


```python
v = np.array([1, 2, 3])
v_col = np.expand_dims(v, axis=1)
v_back = np.squeeze(v_col)

print("v shape:", v.shape)
print("v_col shape:", v_col.shape)
print("v_back shape:", v_back.shape)

assert v_col.shape == (3, 1)
assert np.array_equal(v_back, v)
```

## Edge cases и ловушки

- `reshape` с несовместимым размером вызывает `ValueError`;
- порядок `order='C'` и `order='F'` меняет линейный обход.


```python
try:
    np.arange(10).reshape(3, 4)
except ValueError as e:
    print("reshape error:", e)

m = np.array([[1, 2, 3], [4, 5, 6]])
print("ravel C:", m.ravel(order='C'))
print("ravel F:", m.ravel(order='F'))

assert np.array_equal(m.ravel(order='C'), np.array([1, 2, 3, 4, 5, 6]))
assert np.array_equal(m.ravel(order='F'), np.array([1, 4, 2, 5, 3, 6]))
```
