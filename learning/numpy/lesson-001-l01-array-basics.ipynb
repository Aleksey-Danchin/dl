## 1) Базовые конструкторы: `array`, `zeros`, `ones`, `full`, `empty`

**Что делают**
- `array`: превращает Python-структуру в `ndarray`.
- `zeros` / `ones`: создают массив нулей/единиц.
- `full`: создает массив-константу.
- `empty`: выделяет память без инициализации значений.

**Мини-сигнатуры**
- `np.array(object, dtype=None)`
- `np.zeros(shape, dtype=float)`
- `np.ones(shape, dtype=float)`
- `np.full(shape, fill_value, dtype=None)`
- `np.empty(shape, dtype=float)`

**Ключевые аргументы**
- `object`: исходные данные.
- `shape`: форма массива (кортеж размерностей).
- `dtype`: тип элементов.
- `fill_value`: константа для `full`.

**Когда применять**
- `array` для явных данных;
- `zeros/ones/full` для инициализации;
- `empty` только когда сразу перезаписываешь все элементы.


```python
a = np.array([[1, 2], [3, 4]], dtype=np.int32)
z = np.zeros((2, 3), dtype=np.float32)
o = np.ones((2, 3), dtype=np.int16)
f = np.full((2, 2), -7)
e = np.empty((2, 2))
e[:] = 5

print("a:\n", a, a.dtype)
print("z:\n", z, z.dtype)
print("o:\n", o, o.dtype)
print("f:\n", f, f.dtype)
print("e:\n", e, e.dtype)

assert a.shape == (2, 2)
assert z.shape == (2, 3)
assert np.all(e == 5)
```

## 2) Диагональные и единичные: `eye`, `identity`, `diag`, `diagonal`

**Что делают**
- `eye(N, M=None, k=0)`: 2D массив с единицами на диагонали.
- `identity(n)`: квадратная единичная матрица `n x n`.
- `diag(v, k=0)`: из вектора делает диагональную матрицу (или извлекает диагональ из матрицы).
- `diagonal(a, offset=0)`: извлекает диагональ массива.

**Ключевые аргументы**
- `N`, `M`, `n`: размеры.
- `k`/`offset`: смещение диагонали.
- `v`/`a`: входные данные.


```python
i = np.eye(4)
idn = np.identity(3)
dm = np.diag([2, 4, 6])
vec = np.diagonal(np.array([[9, 1], [5, 7]]))

print("eye 4:\n", i)
print("identity 3:\n", idn)
print("diag from vector:\n", dm)
print("diagonal:", vec)

assert i.shape == (4, 4)
assert np.array_equal(vec, np.array([9, 7]))
```

## 3) Диапазоны и шаблоны: `arange`, `linspace`, `logspace`, `*_like`

**Что делают**
- `arange(start, stop, step)`: равномерная сетка с шагом.
- `linspace(start, stop, num, endpoint=True)`: фиксированное число точек на интервале.
- `logspace(start, stop, num, base=10)`: точки по лог-шкале.
- `zeros_like/ones_like/full_like(a, ...)`: конструкторы по форме и типу шаблона `a`.

**Ключевые аргументы**
- `start`, `stop`, `step`, `num`, `endpoint`, `base`.
- `a`: массив-шаблон для `*_like`.


```python
r = np.arange(-2, 3, 1)
lin = np.linspace(0, 1, 5)
log = np.logspace(1, 3, 3)
base = np.array([[1, 2, 3]], dtype=np.int16)
zl = np.zeros_like(base)
ol = np.ones_like(base)
fl = np.full_like(base, -5)

print("arange:", r)
print("linspace:", lin)
print("logspace:", log)
print("full_like:\n", fl, fl.dtype)

assert np.array_equal(r, np.array([-2, -1, 0, 1, 2]))
assert np.allclose(log, np.array([10., 100., 1000.]))
assert fl.dtype == np.int16
```

## 4) Функциональный конструктор: `fromfunction`

**Сигнатура:** `np.fromfunction(function, shape, *, dtype=float)`  
**Что делает:** вычисляет массив по формуле от индексов.

**Ключевые аргументы**
- `function(i, j, ...)`: функция индексов.
- `shape`: форма результата.
- `dtype`: тип индексов, которые попадут в функцию.

**Когда применять**
Когда проще описать правило генерации значений, чем перечислять значения вручную.


```python
pattern = np.fromfunction(lambda i, j: 10 * i + j, (3, 4), dtype=int)
print(pattern)

assert pattern.shape == (3, 4)
assert pattern[2, 3] == 23
```

## Edge cases и ловушки

- `np.empty` возвращает неинициализированную память;
- `np.arange` с float-шагом может давать неочевидные значения.


```python
e = np.empty((2, 3))
print("empty sample:\n", e)

f = np.arange(0, 1, 0.1)
print("arange float:", f)

assert e.shape == (2, 3)
assert f.shape[0] > 0
```
