## 1. `np.array`


**Сигнатура:** `np.array(object, dtype=None)`  
**Что делает:** превращает Python-структуры в `ndarray`.  
**Когда использовать:** старт почти любой задачи в NumPy.

**Аргументы:**
- `object`: исходные данные (список, кортеж, вложенные структуры).
- `dtype`: целевой тип элементов (опционально).


```python
a = np.array([1, 2, 3])
b = np.array([[1, 2], [3, 4]], dtype=np.int32)
print("a:", a, a.shape, a.dtype)
print("b:\n", b, b.shape, b.dtype)
```

### Задание (`np.array`)

Собери два массива через `np.array`:
- `t_array_1` должен содержать значения `[10, 20, 30]`;
- `t_array_2` должен иметь форму `(2, 3)`, значения `[[1, 2, 3], [4, 5, 6]]` и тип `int64`.


```python
# TODO
t_array_1 = np.array([10, 20, 30])
t_array_2 = np.array([[1, 2, 3], [4, 5, 6]], dtype=np.int64)
```

```python
assert isinstance(t_array_1, np.ndarray)
assert t_array_1.shape == (3,)
assert t_array_2.shape == (2, 3)
assert t_array_2.dtype == np.int64
print("✅ array: ок")
```

## 2. `np.zeros`


**Сигнатура:** `np.zeros(shape, dtype=float)`  
**Что делает:** создает массив нулей заданной формы.

**Аргументы:**
- `shape`: форма массива, например `(3, 2)`.
- `dtype`: тип элементов (по умолчанию `float`).


```python
z = np.zeros((2, 4))
print(z)
print("shape:", z.shape, "dtype:", z.dtype)
```

### Задание (`np.zeros`)

Получи массив `t_zeros`, где все элементы равны нулю, форма `(3, 2)`, тип `float32`.
Используй `np.zeros`.


```python
# TODO
t_zeros = np.zeros((3, 2), dtype=np.float32)
```

```python
assert t_zeros.shape == (3, 2)
assert t_zeros.dtype == np.float32
assert np.all(t_zeros == 0)
print("✅ zeros: ок")
```

## 4. `np.full`


**Сигнатура:** `np.full(shape, fill_value, dtype=None)`  
**Что делает:** заполняет массив одним значением.

**Аргументы:**
- `shape`: форма массива.
- `fill_value`: значение заполнения.
- `dtype`: тип элементов (опционально).


```python
f = np.full((2, 2), 7)
print(f)
```

```python
# TODO
t_full = np.full((3, 3), -2)
```

```python
assert t_full.shape == (3, 3)
assert np.all(t_full == -2)
print("✅ full: ок")
```

## 5. `np.empty`


**Сигнатура:** `np.empty(shape, dtype=float)`  
**Что делает:** выделяет память без инициализации значений.  
**Важно:** использовать только если сразу перезаписываешь все элементы.

**Аргументы:**
- `shape`: форма массива.
- `dtype`: тип элементов.


```python
e = np.empty((2, 2))
print(e)
```

### Задание (`np.empty`)

Создай `t_empty` формы `(2, 3)` через `np.empty`, затем перезапиши массив так,
чтобы в нем не осталось «мусорных» значений: все элементы должны стать равны `5`.


```python
# TODO
t_empty = np.empty((2, 3))
t_empty[...] = 5
```

```python
assert t_empty.shape == (2, 3)
assert np.all(t_empty == 5)
print("✅ empty: ок")
```

## 6. `np.eye`


**Сигнатура:** `np.eye(N, M=None, k=0, dtype=float)`  
**Что делает:** матрица с единицами на диагонали.

**Аргументы:**
- `N`: число строк.
- `M`: число столбцов (если `None`, то равно `N`).
- `k`: сдвиг диагонали.
- `dtype`: тип элементов.


```python
i = np.eye(4, dtype=np.int64)
print(i)
```

### Задание (`np.eye`)

Построй `t_eye` — квадратную матрицу размера `5x5` с единицами на главной диагонали и нулями вне диагонали.
Используй `np.eye`.


```python
# TODO
t_eye = np.eye(5)
```

```python
assert t_eye.shape == (5, 5)
assert np.sum(t_eye) == 5
assert np.array_equal(np.diagonal(t_eye), np.ones(5))
print("✅ eye: ок")
```

## 7. `np.identity`


**Сигнатура:** `np.identity(n, dtype=None)`  
**Что делает:** квадратная единичная матрица `n x n`.

**Аргументы:**
- `n`: размер квадратной матрицы.
- `dtype`: тип элементов (опционально).


```python
ident = np.identity(3)
print(ident)
```

### Задание (`np.identity`)

Построй `t_identity` как единичную матрицу `4x4` через `np.identity`.
Дополнительно получи матрицу тем же размером через `np.eye` и проверь, что результаты совпадают.


```python
# TODO
t_identity = np.identity(4)
t_eye4 = np.eye(4)
```

```python
assert t_identity.shape == (4, 4)
assert np.array_equal(t_identity, t_eye4)
print("✅ identity: ок")
```

## 8. `np.diag`


**Сигнатура:** `np.diag(v, k=0)`  
**Что делает:**
- если `v` — вектор, строит диагональную матрицу;
- если `v` — матрица, извлекает диагональ.

**Аргументы:**
- `v`: вектор или матрица.
- `k`: сдвиг диагонали.


```python
dm = np.diag([1, 3, 5])
print(dm)
```

### Задание (`np.diag`)

Из вектора `[2, 4, 6, 8]` построй матрицу `t_diag_m`, где эти значения стоят на главной диагонали,
а все остальные элементы равны нулю.
Используй `np.diag`.


```python
# TODO
t_diag_m = np.diag([2, 4, 6, 8])
```

```python
assert t_diag_m.shape == (4, 4)
assert np.array_equal(np.diagonal(t_diag_m), np.array([2, 4, 6, 8]))
print("✅ diag: ок")
```

## 9. `np.diagonal`


**Сигнатура:** `np.diagonal(a, offset=0)`  
**Что делает:** извлекает диагональ из матрицы.

**Аргументы:**
- `a`: входная матрица.
- `offset`: сдвиг диагонали.


```python
m = np.array([[9, 1, 2], [3, 8, 4], [5, 6, 7]])
dv = np.diagonal(m)
print(dv)
```

### Задание (`np.diagonal`)

Дана матрица `t_m = np.array([[10, 2], [3, 11]])`.
Извлеки из нее главную диагональ в вектор `t_diag_v` с помощью `np.diagonal`.


```python
# TODO
t_m = np.array([[10, 2], [3, 11]])
t_diag_v = np.diagonal(t_m)
```

```python
assert t_diag_v.shape == (2,)
assert np.array_equal(t_diag_v, np.array([10, 11]))
print("✅ diagonal: ок")
```

## 10. `np.arange`


**Сигнатура:** `np.arange(start, stop, step)`  
**Что делает:** диапазон с фиксированным шагом.

**Аргументы:**
- `start`: начало диапазона.
- `stop`: правая граница (не включается).
- `step`: шаг.


```python
r = np.arange(2, 12, 3)
print(r)
```

### Задание (`np.arange`)

Создай `t_arange`, равный `[-2, -1, 0, 1, 2]`.

```python
assert np.array_equal(t_arange, np.array([-2, -1, 0, 1, 2]))
print("✅ arange: ок")
```

## 11. `np.linspace`


**Сигнатура:** `np.linspace(start, stop, num, endpoint=True)`  
**Что делает:** `num` равномерных точек на линейной шкале.

**Аргументы:**
- `start`, `stop`: границы интервала.
- `num`: количество точек.
- `endpoint`: включать ли правую границу.


```python
lin = np.linspace(0, 1, 5)
print(lin)
```

### Задание (`np.linspace`)

Создай `t_lin` из 5 равномерно распределенных точек на отрезке от `-1` до `1`,
включая оба конца отрезка.
Используй `np.linspace`.


```python
# TODO
t_lin = np.linspace(-1, 1, 5)
t_lin
```

```python
assert t_lin.shape == (5,)
assert np.isclose(t_lin[0], -1)
assert np.isclose(t_lin[-1], 1)
print("✅ linspace: ок")
```

## 12. `np.logspace`


**Сигнатура:** `np.logspace(start, stop, num, base=10.0)`  
**Что делает:** точки по логарифмической шкале (степени `base`).

**Аргументы:**
- `start`, `stop`: степени основания `base`.
- `num`: количество точек.
- `base`: основание логарифмической шкалы.


```python
log = np.logspace(0, 3, 4)
print(log)
```

### Задание (`np.logspace`)

Создай `t_log` как 3 точки логарифмической шкалы по основанию 10,
соответствующие степеням от `10^1` до `10^3`.
Используй `np.logspace`.


```python
# TODO
t_log = np.logspace(1, 3, 3, 10)
t_log
```

```python
assert np.allclose(t_log, np.array([10., 100., 1000.]))
print("✅ logspace: ок")
```

## 13. `np.zeros_like`


**Сигнатура:** `np.zeros_like(a, dtype=None)`  
**Что делает:** нулевой массив формы `a`.

**Аргументы:**

но с нулями во всех ячейках.
Используй `np.zeros_like`.


```python
# TODO
t_base = np.array([[1., 2.], [3., 4.]])
t_zl = np.zeros_like(t_base)
```

```python
assert t_zl.shape == t_base.shape
assert np.all(t_zl == 0)
print("✅ zeros_like: ок")
```
