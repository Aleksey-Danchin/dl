## 1) Семейства `dtype` и память

`dtype` определяет допустимые значения и стоимость хранения.

Ключевые семейства:
- целые: `int8/int16/int32/int64`, `uint8/...`;
- вещественные: `float16/float32/float64`;
- `complex64/complex128`, `bool`, `object`.


```python
arr_i16 = np.array([100, 200, 300], dtype=np.int16)
arr_f64 = np.array([100, 200, 300], dtype=np.float64)

print("int16 bytes:", arr_i16.nbytes, "itemsize:", arr_i16.itemsize)
print("float64 bytes:", arr_f64.nbytes, "itemsize:", arr_f64.itemsize)

assert arr_i16.itemsize == 2
assert arr_f64.itemsize == 8
```

## 2) Приведение типов: `astype`

**Сигнатура:** `arr.astype(dtype, copy=True)`  
Создает массив нового типа.

Важно: `astype(int)` усекает дробную часть, а не округляет.


```python
x = np.array([1.9, -2.1, 3.7], dtype=np.float64)
xi = x.astype(np.int32)
print("x:", x)
print("xi:", xi)

assert np.array_equal(xi, np.array([1, -2, 3], dtype=np.int32))
```

## 3) Числовые границы: `iinfo`, `finfo`

- `np.iinfo(dtype)` -> `min`, `max` для целых;
- `np.finfo(dtype)` -> `eps`, `tiny`, `max` для float.


```python
i8 = np.iinfo(np.int8)
f32 = np.finfo(np.float32)
print("int8 range:", i8.min, i8.max)
print("float32 eps:", f32.eps)

assert i8.min == -128 and i8.max == 127
assert f32.eps > 0
```

## 4) `nan`/`inf` и проверки корректности

- `np.isnan`, `np.isinf`, `np.isfinite`;
- `np.isclose`, `np.allclose` для float-сравнений.


```python
data = np.array([1.0, np.nan, np.inf, -np.inf, 5.0])
print("isnan:", np.isnan(data))
print("isinf:", np.isinf(data))
print("isfinite:", np.isfinite(data))

v1 = 0.1 + 0.2
v2 = 0.3
print("direct eq:", v1 == v2)
print("isclose:", np.isclose(v1, v2))

assert np.isnan(data).sum() == 1
assert np.isinf(data).sum() == 2
assert np.isfinite(data).sum() == 2
assert np.isclose(v1, v2)
```

## Edge cases и ловушки

- переполнение fixed-size int (`uint8`);
- ложные сравнения float через `==`.


```python
u = np.array([255], dtype=np.uint8)
print("overflow uint8:", u + 1)

assert (u + 1)[0] == 0
assert not ((0.1 + 0.2) == 0.3)
assert np.isclose(0.1 + 0.2, 0.3)
```
