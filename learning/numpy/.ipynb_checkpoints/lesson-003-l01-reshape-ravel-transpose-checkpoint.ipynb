## 1) `reshape`: новая форма при том же количестве элементов

**Что делает**  
`reshape` возвращает массив с новой формой, если число элементов совместимо.

**Минимальные сигнатуры**
- `arr.reshape(shape, order='C')`
- `np.reshape(a, newshape, order='C')`

**Ключевые параметры**
- `shape` / `newshape`: целевая форма (кортеж).
- `order`: порядок чтения/записи (`'C'` построчно, `'F'` по столбцам).

**Ожидаемый результат**
`size` сохраняется, меняется только `shape`.

**Когда применять**
Когда нужно подготовить данные под матричные операции или пакетную обработку.


```python
x = np.arange(12)
x2 = x.reshape(3, 4)

print("x:", x, x.shape)
print("x2:\n", x2, x2.shape)

assert x.size == x2.size
assert x2.shape == (3, 4)
```

### Разбор вывода
`reshape` не создает новые значения: те же 12 чисел просто раскладываются по сетке `3x4`.

## 2) `transpose` и `.T`: перестановка осей

**Что делает**
- `transpose` переставляет оси массива;
- `.T` — короткая запись транспонирования (особенно для 2D).

**Минимальные сигнатуры**
- `np.transpose(a, axes=None)`
- `arr.transpose(*axes)`
- `arr.T`

**Ключевые параметры**
- `axes`: новый порядок осей; если `None`, оси разворачиваются в обратном порядке.

**Ожидаемый результат**
У 2D массива форма `(m, n)` становится `(n, m)`.

**Когда применять**
При подготовке матриц для линейной алгебры и согласовании форм перед операциями.


```python
m = np.arange(1, 7).reshape(2, 3)
mt = m.T

print("m:\n", m)
print("mt:\n", mt)

assert m.shape == (2, 3)
assert mt.shape == (3, 2)
assert np.array_equal(mt, np.transpose(m))
```

### Разбор вывода
Транспонирование "меняет местами" строки и столбцы: элемент `[i, j]` переходит в `[j, i]`.

## 3) `ravel` vs `flatten`: представление или копия

**Что делает**
- `ravel` пытается вернуть плоское представление (view), когда это возможно;
- `flatten` возвращает новый плоский массив (обычно копию).

**Минимальные сигнатуры**
- `np.ravel(a, order='C')`
- `arr.ravel(order='C')`
- `arr.flatten(order='C')`

**Ключевой параметр**
- `order`: порядок обхода (`'C'`, `'F'`, ...).

**Ожидаемый результат**
Оба дают форму `(N,)`, но поведение при изменении данных может отличаться.

**Когда применять**
- `ravel` — когда важна производительность и можно работать с view;
- `flatten` — когда нужна изолированная копия.


```python
a = np.arange(6).reshape(2, 3)
av = a.ravel()
ac = a.flatten()

a[0, 0] = 999

print("a:\n", a)
print("av (ravel):", av)
print("ac (flatten):", ac)

assert av[0] == 999
assert ac[0] == 0
```

### Разбор вывода
После изменения исходного массива `ravel` отражает изменение, а `flatten` сохраняет старое значение.
Это главный практический признак "view vs copy" в типичном случае.

## 4) Комбинирование операций: `reshape -> transpose -> ravel`

**Что делает**
Последовательный пайплайн позволяет гибко перестроить представление данных под нужный формат выхода.

**Минимальная идея пайплайна**
1. построить диапазон чисел;
2. задать двумерную форму;
3. переставить оси;
4. получить одномерный результат.

**Когда применять**
При подготовке данных для сериализации, логирования, или нестандартного входного формата модели.


```python
p = np.arange(12)
p1 = p.reshape(3, 4)
p2 = p1.T
p3 = p2.ravel()

print("p1 shape:", p1.shape)
print("p2 shape:", p2.shape)
print("p3 shape:", p3.shape)
print("p3:", p3)

assert p3.shape == (12,)
```

### Разбор вывода
Каждый шаг меняет геометрию данных. Итоговый вектор содержит те же элементы,
но порядок может отличаться от исходного `np.arange`.

## Edge cases и ловушки

1) `reshape` падает с ошибкой, если размер несовместим.  
2) Порядок обхода (`order='C'` vs `order='F'`) влияет на вид плоского результата.


```python
# Ловушка 1: несовместимая форма
try:
    bad = np.arange(10).reshape(3, 4)
except ValueError as e:
    print("reshape error:", e)

# Ловушка 2: разный порядок ravel
m = np.array([[1, 2, 3], [4, 5, 6]])
flat_c = m.ravel(order='C')
flat_f = m.ravel(order='F')

print("flat C:", flat_c)
print("flat F:", flat_f)

assert np.array_equal(flat_c, np.array([1, 2, 3, 4, 5, 6]))
assert np.array_equal(flat_f, np.array([1, 4, 2, 5, 3, 6]))
```

## Шпаргалка модуля

- `reshape`: новая форма при сохранении `size`.
- `transpose` / `.T`: перестановка осей.
- `ravel`: плоское представление (часто view).
- `flatten`: плоская копия.
- `order='C'/'F'`: порядок чтения элементов при разворачивании.
