## Блок 1 - Основные dtypes в NumPy

### Теория

Массивы NumPy однородны: один массив -> один dtype. Основные семейства типов: целые (`int32`, `int64`), вещественные (`float32`, `float64`), логические (`bool`) и комплексные (`complex64`, `complex128`).

```python
import numpy as np

a_int = np.array([1, 2, 3], dtype=np.int32)
a_float = np.array([1.0, 2.5, 3.1], dtype=np.float64)
a_bool = np.array([True, False, True], dtype=np.bool_)
a_complex = np.array([1 + 2j, 3 + 0j], dtype=np.complex128)

for name, arr in [
    ("int", a_int),
    ("float", a_float),
    ("bool", a_bool),
    ("complex", a_complex),
]:
    print(name, arr.dtype, "itemsize=", arr.itemsize, "bytes")
```

```python
# Самопроверка примера из блока 1
assert a_int.dtype == np.int32
assert a_float.dtype == np.float64
assert a_bool.dtype == np.bool_
assert a_complex.dtype == np.complex128
print("Проверки примера блока 1 пройдены")
```

## Блок 2 - Приведение типов и точность

### Теория

Приведение типов (`astype`) может терять информацию. При переводе из float в int дробная часть отбрасывается. Меньшие dtypes могут снижать точность.

```python
vals = np.array([1.9, 2.1, -3.7], dtype=np.float64)
vals_i32 = vals.astype(np.int32)

print("original:", vals, vals.dtype)
print("casted :", vals_i32, vals_i32.dtype)

tiny = np.array([1.123456789], dtype=np.float64)
tiny32 = tiny.astype(np.float32)
print("float64:", tiny[0])
print("float32:", tiny32[0])
```

```python
# Самопроверка примера из блока 2
assert vals_i32.dtype == np.int32
assert np.array_equal(vals_i32, np.array([1, 2, -3], dtype=np.int32))
assert tiny32.dtype == np.float32
print("Проверки примера блока 2 пройдены")
```

## Блок 3 - Память и интуиция производительности

### Теория

Память — один из ключевых факторов производительности. Для больших массивов правильный выбор `dtype` снижает нагрузку на RAM и ускоряет последующие вычисления.

```python
n = 1_000_000
arr32 = np.ones(n, dtype=np.float32)
arr64 = np.ones(n, dtype=np.float64)

print("float32 nbytes:", arr32.nbytes)
print("float64 nbytes:", arr64.nbytes)
print("ratio float64/float32:", arr64.nbytes / arr32.nbytes)
```

```python
# Самопроверка примера из блока 3
assert arr32.itemsize == 4
assert arr64.itemsize == 8
assert arr64.nbytes == 2 * arr32.nbytes
print("Проверки примера блока 3 пройдены")
```
